# Scrapping the data from Open Alex

_Foreword_

The goal of this notebook is to do a request to OpenAlex to download all the data related to our list of technologies.

Importing the modules I will need.

```python
import requests
from collections import defaultdict
import time
import json
from tqdm import tqdm
```

Defining some auxiliary function I will use to scrap the data.

```python
def get_works_from_response(r_json):
    """
    Get the results from response 
    """
    return r_json['results']
```

```python
def get_new_cursor_from_results(r_json):
    """
    Need to get new cursor from response to advance and get the following papers
    """
    return r_json['meta']['next_cursor']
```

Defining some important variables.

```python
url = 'https://api.openalex.org/'

type_ = 'works'
```

```python
# need this to end up in "good user pool"
headers = {
    'User-Agent': "alessandro.tavazzi@epfl.ch"  # your e-mail address
}
```

Below are the technologies with their OpenAlex id. They were chosen with the Delphi method. Note, that not all of them are available on OpenAlex.

```python
# technologies from Sarah's work
concept_ids = {
    'Authentication protocol': 'C21564112',
    'Biometrics': 'C184297639',
    'Blockchain': 'C2779687700',
    # 'Database Encryption': '',
    'Differential Privacy': 'C23130292',
    'Digital rights management': 'C537843408',
    'Digital signature': 'C118463975',
    'Disk Encryption': 'C9368797',
    'Distributed algorithm': 'C130120984',
    'Electronic voting': 'C2780612046',
    # 'Email encryption': '',
    'Functional encryption': 'C2780746774',
    'Hardware acceleration': 'C13164978',
    'Hardware security module': 'C39217717',
    'Hash function': 'C99138194',
    'Homomorphic encryption': 'C158338273',
    'Identity management': 'C555379026',
    # 'Identity-based encryption': '',
    'Key management': 'C17886624',
    'Link encryption': 'C69254412',
    'Post-quantum cryptography': 'C108277079',
    # 'Private set intersection': '',
    'Public-key cryptography': 'C203062551',
    'Quantum key distribution': 'C95466800',
    'Quantum cryptography': 'C144901912',
    'Random number generation': 'C201866948',
    # 'Searchable symmetric encryption': '',
    'Symmetric-key algorithm': 'C65302260',
    'Threshold cryptosystem': 'C123744220',
    'Trusted Computing': 'C2776831232',
    # 'Trusted execution environment': '',
    'Tunneling protocol': 'C76885553',
    'Zero-knowlegde proof': 'C176329583'
}
```

Creating a dictionary to store all my data.

```python
raw_data = defaultdict(list)
```

```python
# https://docs.openalex.org/about-the-data/concept
payload = {
    'cursor': '*',
    'filter': 'concept.id:C2778407487',
    'per-page': 200,
}
```

Now I iterate over the lists to do the requests which give me the data.

```python
# iterate through all concepts
for concept, concept_id in tqdm(concept_ids.items()):
    if concept_id == '':
        continue
    payload['cursor'] = '*'
    payload['filter'] = f'concept.id:{concept_id}'
    limit_counter = 0
    # get all works for concept
    while True:
        if limit_counter == 0:
            start_time = time.time()
        response = requests.get(url + type_, headers=headers, params=payload)
        print(response)
        r_json = response.json()
        if len(r_json['results']) == 0:
            break
        raw_data[concept].extend(get_works_from_response(r_json))
        payload['cursor'] = get_new_cursor_from_results(r_json)
        limit_counter += 1

        if limit_counter == 10:
            # only 10 requests in 1 second
            end_time = time.time()
            delta = end_time - start_time
            if delta < 1:
                print(f'Sleeping: {delta}')
                time.sleep(delta)
            limit_counter = 0
```

I save my data in the appropriate folder.

```python
with open('data_creation_variables/raw_data.json', 'w') as f:
    json.dump(raw_data, f)
```
